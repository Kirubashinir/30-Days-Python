1️⃣ What is a String?

A string is a sequence of characters enclosed in quotes.

```python
name = "Kirubashini"
language = 'Python'

print(name)
print(language)
```

2️⃣ String Indexing

Each character has an index position.

```python
text = "Python"

print(text[0])  # P
print(text[1])  # y
print(text[-1]) # n
```

3️⃣ String Slicing

Extract a portion of a string.

```python
text = "Python Programming"

print(text[0:6])
print(text[7:18])
print(text[:6])
print(text[-11:])
```

4️⃣ String Operations

```python
first = "Hello"
second = "World"

print(first + " " + second)
print(first * 3)
```

5️⃣ Useful String Methods
Uppercase

```python
##Uppercase
name = "python"

print(name.upper())

##Lowercase
name = "PYTHON"

print(name.lower())


##Capitalize
name = "python"

print(name.capitalize())

##Replace
text = "I love Java"

print(text.replace("Java", "Python"))

##Length
text = "Machine Learning"

print(len(text))
```

6️⃣ Checking Strings

```python
text = "Python Programming"

print("Python" in text)
print("Java" in text)

print(text.startswith("Python"))
print(text.endswith("Programming"))
```

7️⃣ String Formatting

```python
##f-Strings (Recommended)
name = "Kirubashini"
role = "AI Engineer"

print(f"My name is {name} and I want to become an {role}.")

##format()
name = "Kirubashini"

print("Welcome {}".format(name))
```

8️⃣ Escape Characters

```python
print("Hello\nWorld")
print("Python\tProgramming")
print("He said \"Hello\"")
```

🚀 Practice Programs
Program 1: Count Characters

```python
text = input("Enter a string: ")

print("Length:", len(text))
```

Program 2: Reverse a String

```python
text = input("Enter a string: ")

print("Reversed:", text[::-1])
```

Program 3: Palindrome Checker

```python
text = input("Enter a word: ")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

Program 4: Count Vowels

```python
text = input("Enter a string: ")

count = 0

for char in text.lower():
    if char in "aeiou":
        count += 1

print("Vowels:", count)
```
