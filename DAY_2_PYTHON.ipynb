1️⃣ Arithmetic Operators

Used for mathematical calculations.

Operator	Description	Example

+	Addition	5 + 3



-	Subtraction	5 - 3

*	Multiplication	5 * 3

/	Division	5 / 3

//	Floor Division	5 // 3

%	Modulus	5 % 3

**	Exponent	5 ** 3

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

2️⃣ Comparison Operators

Used to compare values.

Operator	Meaning

* ==	Equal to

* !=	Not Equal to

* >	Greater than

* <	Less than

* >=	Greater than or Equal

* <=	Less than or Equal

```python
x = 20
y = 10

print(x == y)
print(x != y)
print(x > y)
print(x < y)
```

3️⃣ Assignment Operators

Assign values to variables.

```python
x = 10

x += 5
print(x)

x -= 2
print(x)

x *= 3
print(x)

x /= 2
print(x)
```

4️⃣ Logical Operators

Used to combine conditions.

Operator	Meaning
and	Both conditions True
or	At least one True
not	Reverse condition

```python
age = 22
has_id = True

print(age >= 18 and has_id)
print(age >= 25 or has_id)
print(not has_id)
```

5️⃣ Membership Operators

Check if a value exists in a sequence.


```python
skills = ["Python", "SQL", "Machine Learning"]

print("Python" in skills)
print("Java" in skills)
print("Java" not in skills)
```

6️⃣ Identity Operators

Check whether objects share the same memory location.

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)
print(a is c)

print(a is not c)
```

Program 2: Even or Odd

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

Program 3: Largest Number

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print(a, "is largest")
else:
    print(b, "is largest")
```
